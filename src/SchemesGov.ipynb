```python
import pandas as pd
import plotly.express as px
import dash 
import dash_html_components as html 
import dash_core_components as dcc
import plotly.graph_objects as go
from dash.dependencies import Input, Output
global x
x=0

app = dash.Dash()

fig = go.Figure()
##fig.update_layout(plot_bgcolor='blue')
fig.update_yaxes(showgrid=False) # disable the horizontal grid lines
fig.update_xaxes(showgrid=False)


app.layout=html.Div([html.Div(html.H1('Details of Goverment Schemes',style={'color':'white',            #all should be as string
                                                   'text-align':'center',
                                                   'background-color':'gray',
                                                   'border-style': 'double',
                                                   'display':'inline-block',  # display list items horizontally instead of vertically.
                                                   'width':'100%'
                                                    })),
                   
                   
                   # syle for html.div
                    html.Div(dcc.Dropdown(id='dropdown1',options=[{'label': 'Ministry of Agriculture and Farmers Welfare','value': 'ag'},
                                                                  {'label': 'Sarva Shiksha Abhiyan','value': 'SSA'}],
                                                                  placeholder="Select an Mission/Scheme/Ministry"),
                                                                     
                                                                  style={'width':'30%','display':'inline-block'}),
                    html.Div(dcc.Dropdown(id='dropdown2',options=[{'label': '2014-15 - Actual Exp.','value': 'Yr14'},
                                                                  {'label': '2015-16 - Actual Exp.','value': 'Yr15'},
                                                                  {'label': '2016-17 - Actual Exp.','value': 'Yr16'}],
                                                                  placeholder="Select a Time Period"),                                                                  
                                                                  style={'width':'30%','display':'inline-block'}),
                    html.Div(html.Button('Submit', id='button',n_clicks=0)),
               
                    html.Div(dcc.Graph(id='firstgraph1',figure=fig,style={'width':'100%','display':'inline-block'}))
                    ])
                     
@app.callback(
    Output('firstgraph1', 'figure'),
    [Input('dropdown1', 'value'),Input('dropdown2', 'value'),Input('button', 'n_clicks'),]) 

def update_figure(dropdown1,dropdown2,n_clicks):
    global x
    global fig
    if (dropdown1=='ag') & (dropdown2=='Yr14') & (n_clicks>x) :
        fig= go.Figure(data=go.Bar(x=DeptOfAgri['Name of Mission/Scheme'],y=DeptOfAgri['2014-15 - Actual Exp.'],text=DeptOfAgri['2014-15 - Actual Exp.'],textposition='auto'))
        fig.update_layout(title="Ministry of Agriculture and Farmers Welfare (2014-15 - Actual Exp.)")

    if (dropdown1=='ag') & (dropdown2=='Yr15') & (n_clicks>x) :
        fig= go.Figure(data=go.Bar(x=DeptOfAgri['Name of Mission/Scheme'],y=DeptOfAgri['2015-16 - Actual Exp.'],text=DeptOfAgri['2015-16 - Actual Exp.'],textposition='auto'))
        fig.update_layout(title="Ministry of Agriculture and Farmers Welfare (2015-16 - Actual Exp.)")
  
        
    if (dropdown1=='ag') & (dropdown2=='Yr16') & (n_clicks>x):
        fig=go.Figure(go.Bar(x=DeptOfAgri['Name of Mission/Scheme'],y=DeptOfAgri['2016-17 - Actual Exp.'],text=DeptOfAgri['2016-17 - Actual Exp.'],textposition='auto'))
        fig.update_layout(title="Ministry of Agriculture and Farmers Welfare (2016-17 - Actual Exp.)")
    
    
    x=n_clicks
    return fig


if __name__ == '__main__':        
    app.run_server()
```
